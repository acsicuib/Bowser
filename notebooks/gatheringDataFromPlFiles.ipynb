```python
import pandas as pd
import numpy as np
import glob
from subprocess import Popen, PIPE, TimeoutExpired
import os
import re
```

```python
plfiles = "../results/results_SMALL_OSMOTIC_SEED_vNew_X/models_9/"
df = pd.DataFrame(columns=["des","node","time","action","flavour","file"])
for abs_name in glob.glob(plfiles+"*.pl"):
    print(abs_name)
    name = os.path.basename(abs_name)
    parts = name.split("_")
    node = int(parts[2].replace("n",""))
    des = int(parts[3].replace("DES",""))
    tries = int(parts[4])
    time =  int(parts[5].replace(".pl",""))
    bloc = int(parts[1].replace("B",""))
    
    
    cmd = ["swipl", abs_name, "s"+str(des)]

    # print(cmd)
    p = Popen(cmd, stdout=PIPE, stderr=PIPE)
    stdout, stderr = p.communicate(timeout=10)

    expr = stdout.decode("utf-8")
    expr = expr.replace("\n","")

    it = iter("".join(c for c in expr if c not in "()[] ").split(","))
    result = [(x, y, z, v ) for x, y, z, v in zip(it, it, it, it)]

    if len(result)>0:
        df.loc[len(df)]=[des,node,time,result[0][0],result[0][3],name]
    else:
        df.loc[len(df)]=[des,node,time,"None","None",name]

```

```python
df = df.sort_values(["node","time"])
```

```python
columns = ["file",
           "app",
           "cf", # current flavour
           "avaiHW", # current available HW
           "degree", # neighbours or degree node 
           "req_channels", # number of requests
           "diff_req_channels", # different number of nodes who send requests 
           "req_sum", #number of requests
           "sum_lat","avg_lat","std_lat","max_lat","min_lat",#Requests latency
           "f1","f1_HW","f1_MaxReq","f2","f2_HW","f2_MaxReq","f3","f3_HW","f3_MaxReq", #App flavours
           "app_max_lat" #app max latency supported 
          ]
    
             
dfacts = pd.DataFrame(columns=columns)

rRequest = re.compile("requests\(s\d+,\[(n\d+)|self],\d+,\d+\).\n")
rNode = re.compile("node\(self,\d+")
rServiceInstance = re.compile("serviceInstance\(s")

map_services = {"s":1,"m":2,"l":3}
services = dict()
services[1]=[1,1,10,2,2,20,3,5,50,10]
services[2]=[1,1,10,2,2,20,3,5,50,10]
services[3]=[1,0,0,2,4,20,3,0,0,10]

for ix,val in df.iterrows():
    setOfNeigh = set()
    totalRequests = []
    latencyOfRequests = []
    
    nodeAvailHW= []
    nodeNeigh = []
    try:
        with open(plfiles+val.file) as f:
            for line in f:
                #requests
                result = rRequest.search(line)
                if result:
                    line = line.replace("requests(s","").replace(").\n","").split(",")
                    
                    totalRequests.append(int(line[2]))
                    latencyOfRequests.append(int(line[3]))
                    
                    if "self" in line[1]:
                        setOfNeigh.add(-1) #requests from the same node
                    else:
                        setOfNeigh.add(int(line[1].replace("[n","").replace("]","")))
                else:
                    #current node
                    result = rNode.search(line)
                    if result:
                        line = line.replace("node(self","").replace(").\n","").split(",")
                        nodeAvailHW = int(line[1])
                        nodeNeigh = line[2].replace("n","").strip('][').split(', ')
                    else:
                        #serviceInstance fact
                        result = rServiceInstance.search(line)
                        if result:
                            line = line.replace("serviceInstance(s","").replace(").\n","").split(",")
                            app = int(line[1].replace("app",""))

                            fl = line[2].replace("(","")[0]
                            current_flavour = map_services[fl]

        latencyOfRequests = np.array(latencyOfRequests)
        totalRequests = np.array(totalRequests)
        sample = [val.file,app,current_flavour,nodeAvailHW,
                  len(nodeNeigh), #Degree node  
                  len(totalRequests), # number of requests
                  len(setOfNeigh), # different number of nodes who send requests 
                  totalRequests.sum(), #number of requests
                  latencyOfRequests.sum(),latencyOfRequests.mean(),latencyOfRequests.std(),latencyOfRequests.max(),latencyOfRequests.min(),
                 ]
        
        
        sample = sample+services[app]
        dfacts.loc[len(dfacts)]=sample
    except:
        print(val.file)
        print("App", app)
        print("Currenf flavour", current_flavour)

        print("Number of incomming requests",len(totalRequests))
        print("Total number of requests",sum(totalRequests))

        print("Node available HW", nodeAvailHW)
        print("Number of neigh ",len(nodeNeigh))

        break
```
