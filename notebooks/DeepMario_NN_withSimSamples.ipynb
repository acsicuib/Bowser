```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

from fastai.tabular.data import TabularDataLoaders
from fastai.tabular.core import Categorify
from fastai.tabular.learner import TabularLearner
from fastai.tabular.model import TabularModel
from fastai.metrics import accuracy
#https://medium.com/analytics-vidhya/neural-networks-for-tabular-data-touching-the-untouched-efd66c950463
```

```python
path = "../results/results_SmallFog_X/samples.csv"
df = pd.read_csv("../results/results_SmallFog_X/samples.csv")
print(df.head())
samples = len(df)
print(samples)
```

```python
df = df.drop(columns=["sim"])
print(df.columns)
```

```python
dep_var = 'fit'
cat_names = ['currentFlavour', 'action']
cont_names = [x for x in df.columns if x not in cat_names]
procs = [Categorify]
```

```python
dls = TabularDataLoaders.from_df(df, path=path, 
                                 cat_names=cat_names, 
                                 cont_names=cont_names, procs=procs,
                                 y_names=dep_var,
                                 valid_idx=list(range(800,samples)))
```

```python
dls.show_batch()
```

```python
emb_szs = [(4,2), (17,8)]
m = TabularModel(emb_szs, n_cont=28, out_sz=1, layers=[200,100]).eval()
```

```python

learn = TabularLearner(dls, model = m,metrics=accuracy)
learn.fit(5, 1e-2)
#epoch, train_loss, valid_loss, accuracy, time
```
